```python
from langchain_text_splitters import (
    RecursiveCharacterTextSplitter,
    Language
)

GAME_CODE="""
<!DOCTYPE html>
<html lang="zh-CN">
<head>
<meta charset="UTF-8" />
<meta name="viewport" content="width=device-width, initial-scale=1.0" />
<title>圆圈内弹珠反弹</title>
<style>
body{margin:0;display:flex;justify-content:center;align-items:center;height:100vh;background:#111;color:#fff;font-family:sans-serif}
canvas{background:#000;border-radius:12px;box-shadow:0 0 20px rgba(255,255,255,.15)}
</style>
</head>
<body>
<canvas id="c" width="800" height="800"></canvas>
<script>
const canvas=document.getElementById('c');
const ctx=canvas.getContext('2d');
const W=canvas.width,H=canvas.height;
const cx=W/2, cy=H/2, R=300;
const ball={x:cx+120,y:cy-80,vx:3.8,vy:2.9,r:14};
function update(){
  ball.x+=ball.vx; ball.y+=ball.vy;
  const dx=ball.x-cx, dy=ball.y-cy;
  const dist=Math.hypot(dx,dy);
  if(dist+ball.r>=R){
    const nx=dx/dist, ny=dy/dist;
    const dot=ball.vx*nx+ball.vy*ny;
    ball.vx=ball.vx-2*dot*nx;
    ball.vy=ball.vy-2*dot*ny;
    const overlap=dist+ball.r-R;
    ball.x-=nx*overlap;
    ball.y-=ny*overlap;
  }
}
function draw(){
  ctx.clearRect(0,0,W,H);
  ctx.strokeStyle='white'; ctx.lineWidth=3;
  ctx.beginPath(); ctx.arc(cx,cy,R,0,Math.PI*2); ctx.stroke();
  const g=ctx.createRadialGradient(ball.x-4,ball.y-4,2,ball.x,ball.y,ball.r);
  g.addColorStop(0,'#fff'); g.addColorStop(1,'#4da3ff');
  ctx.fillStyle=g;
  ctx.beginPath(); ctx.arc(ball.x,ball.y,ball.r,0,Math.PI*2); ctx.fill();
}
function loop(){update();draw();requestAnimationFrame(loop);} loop();
</script>
</body>
</html>
"""

python_splitter=RecursiveCharacterTextSplitter.from_language(
    Language.HTML,
    chunk_size=100,
    chunk_overlap=0
)

python_docs=python_splitter.create_documents([GAME_CODE])
python_docs
```
