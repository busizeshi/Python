## SDK调用

````python
from openai import OpenAI

client = OpenAI(
    api_key=QWEN_API_KEY,
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"
)

resp = client.chat.completions.create(
    model="qwen-turbo",
    messages=[
        {"role": "user", "content": "解释LangChain的作用。"},
    ]
)

print(resp.choices[0].message.content)
````

## LangChain集成

````python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatOpenAI(
      model="qwen-turbo",
      api_key=QWEN_API_KEY,
      base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
  )

prompt = ChatPromptTemplate.from_messages([
      ("system", "你是一个精通 LangChain 的专家，使用中文回答。"),
      ("human", "{question}")
  ])

chain = prompt | llm
resp = chain.invoke({"question": "Explain the purpose of LangChain."})
print(resp.content)
````

# 工具调用与模型上下文协议

## 概要

### 工具调用 (Tool Calling / Function Calling)
工具调用并非模型真的去“点击”了某个按钮，而是一种结构化交互机制。

__核心原理__
- 定义阶段：开发者用 JSON Schema 描述一个函数的名称、功能及参数。例如：get_weather(city: string)。

- 推理阶段：用户提问（如“杭州今天多少度？”），模型判断现有知识无法回答，于是返回一个特定格式的字符串（通常是 JSON），表明它想调用哪个工具以及参数是什么。

- 执行阶段：代码环境（你的后端程序）拦截到这个请求，实际运行该函数，并获取结果（如 25°C）。

- 回复阶段：将结果喂回给模型，模型再组织成自然语言告诉用户。

__为什么它很重要?__

消除幻觉：让模型访问实时数据。

打破黑盒：允许模型操作外部系统（如发邮件、查数据库）。

结构化输出：强制模型按预设格式输出，便于程序自动化处理。

### 模型上下文协议 (Model Context Protocol, MCP)
MCP 是由 Anthropic 在 2024 年底推出的一个开放标准。如果说“工具调用”是具体的招式，那么 MCP 就是连接插座的统一标准。

__核心痛点__

以前，如果你想让 Claude、ChatGPT 和 Gemini 都能访问你的 Google Drive，你得为每个模型写一套接入逻辑。而 MCP 的目标是：一次编写，到处连接。

__MCP 的三层架构__

MCP 采用了类似于 LSP（编程语言服务器协议）的架构：

MCP Hosts：使用数据的平台（如 Claude Desktop、IDE、各种 Agent 框架）。

MCP Clients：Host 内部负责与服务器通信的客户端。

MCP Servers：暴露具体能力的服务器（如提供 GitHub 搜索、本地文件读取或数据库查询的服务）。

__MCP 的三大支柱内容__

Resources（资源）：类似于只读文件。模型可以读取数据库 Schema、本地文档等背景信息。

Tools（工具）：可执行的操作。这对应了我们上面提到的 Tool Calling，但在 MCP 下它是标准化的。

Prompts（提示词模板）：预设的代码审查、翻译等交互模版。

## 如何将工具挂载到智能体中

- 工具注册：开发者提供工具的描述文件（如函数名、输入参数、返回值类型）
- 协议对接：通过MCP将工具描述转化为智能体可以识别的格式
- 上下文注入：在对话或任务执行过程中，系统会将工具能力注入大语言模型的上下文，使其在规划时能感知到这些可调用资源
- 调用与执行：当大语言模型认为某个任务需要调用工具时，它会生成符合MCP的调用指令，由执行引擎路由到对应工具，执行后再将结果返回模型

### 构建MCP Server

构建一个完整的 MCP (Model Context Protocol) Demo，最快且最主流的方式是使用 Python。Anthropic 官方提供了一个非常类似于 FastAPI 的高级封装 —— FastMCP

MCP Server 的核心任务是暴露资源 (Resources) 或 工具 (Tools)

```python
from mcp.server.fastmcp import FastMCP

# 1. 初始化 FastMCP Server
# 这里的名称会在 Client 连接时作为标识
mcp = FastMCP("MyDemoServer")

# 2. 定义一个 Tool (工具调用)
# 使用 @mcp.tool() 装饰器，SDK 会自动提取函数名、参数类型声明和 docstring 转化为 JSON Schema
@mcp.tool()
def fetch_user_data(user_id: int) -> str:
    """根据 User ID 获取用户的模拟数据库信息"""
    # 在真实业务中，这里可以是你的 MySQL 或 Vector DB 查询逻辑
    mock_db = {
        1: {"name": "Alice", "role": "Admin", "status": "Active"},
        2: {"name": "Bob", "role": "Developer", "status": "Inactive"}
    }

    user = mock_db.get(user_id)
    if user:
        return f"User Found: {user['name']}, Role: {user['role']}"
    return "Error: User not found."

@mcp.tool()
def calculate_system_load(cores: int, active_threads: int) -> float:
    """模拟计算当前系统的负载率"""
    if cores <= 0:
        return 0.0
    return round((active_threads / (cores * 2)) * 100, 2)

if __name__ == "__main__":
    # 3. 运行 Server
    # 默认使用 stdio (标准输入输出) 进行进程间通信，这是 MCP 最常用的本地交互方式
    print("Starting MCP Server...", file=sys.stderr) # 注意：日志必须打印到 stderr，因为 stdout 被用作协议通信
    mcp.run()
```

### 构建 MCP Client

__两种调用方式__

1. 纯 MCP Client 调用 (底层执行)：完全靠代码硬编码，人或程序决定什么时候调用什么工具。

2. 结合 OpenAI Function Calling (上层智能)：将 MCP Server 提供的工具动态转换给大模型，让大模型（如 GPT-4o）来决定何时调用、传入什么参数，最后通过 MCP Client 执行。

#### 纯代码使用MCP Client调用

这种方式不依赖任何大模型，纯粹是把 MCP 当作一个高级的、标准化的 RPC（远程过程调用）协议来使用。

__核心逻辑__：启动 Server -> 获取工具列表 -> 通过代码指定要调用的工具和参数 -> 执行并获取结果。

```
import asyncio
from mcp import ClientSession, StdioServerParameters
from mcp.client.stdio import stdio_client

async def main():
    # 1. 配置 Server 启动参数
    # 这里告诉 Client 去运行我们刚才写的 server.py
    server_params = StdioServerParameters(
        command="python",
        args=["server.py"]
    )

    print("🔌 正在连接到 MCP Server...")

    # 2. 通过 stdio 建立连接
    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:
            # 3. 初始化协议 (握手)
            await session.initialize()
            print("✅ 握手成功！")

            # 4. 获取 Server 提供的工具列表
            tools_response = await session.list_tools()
            print("\n🛠️ Server 提供的可用工具:")
            for tool in tools_response.tools:
                print(f" - {tool.name}: {tool.description}")

            # 5. 调用指定的工具 (Tool Calling)
            print("\n🚀 开始调用工具: fetch_user_data(user_id=1)")
            result1 = await session.call_tool(
                name="fetch_user_data",
                arguments={"user_id": 1}
            )
            # 解析并打印结果
            # result.content 通常是一个列表，包含文本或图像等多种模态的数据
            print(f"📥 结果: {result1.content[0].text}")

            print("\n🚀 开始调用工具: calculate_system_load(cores=8, active_threads=12)")
            result2 = await session.call_tool(
                name="calculate_system_load",
                arguments={"cores": 8, "active_threads": 12}
            )
            print(f"📥 结果: {result2.content[0].text}%")

if __name__ == "__main__":
    # 运行异步客户端
    asyncio.run(main())
```

#### 结合 OpenAI Client 实现 Agent 调用 (动态编排)

这是真正的 AI Agent 玩法。我们将 MCP 协议中的工具（JSON Schema）无缝桥接到 OpenAI 的 Tools 中。

__核心逻辑__：

- 从 MCP Server 获取工具定义。

- 将 MCP 工具定义转换为 OpenAI 识别的 JSON 格式。

- 把用户的自然语言问题（如“帮我查一下1号用户的状态”）发给 OpenAI。

- OpenAI 返回想要调用的工具名称和参数（Function Calling）。

- 拦截该请求，交由 MCP Client 实际执行该工具。

- 把执行结果喂给 OpenAI，生成最终的人类语言回复。

```
import asyncio
import json
from openai import AsyncOpenAI
from mcp import ClientSession, StdioServerParameters
from mcp.client.stdio import stdio_client

# 初始化 OpenAI 客户端 (请确保环境中配置了 OPENAI_API_KEY)
llm_client = AsyncOpenAI()

async def llm_mcp_agent():
    # 1. 启动并连接 MCP Server
    server_params = StdioServerParameters(command="python", args=["server.py"])

    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:
            await session.initialize()

            # 2. 从 MCP Server 获取可用工具
            mcp_tools_response = await session.list_tools()

            # 3. 将 MCP 工具转换为 OpenAI 的 tools 格式
            openai_tools = []
            for tool in mcp_tools_response.tools:
                openai_tools.append({
                    "type": "function",
                    "function": {
                        "name": tool.name,
                        "description": tool.description,
                        "parameters": tool.inputSchema  # MCP的Schema完美兼容OpenAI
                    }
                })

            # 4. 用户输入自然语言
            messages = [{"role": "user", "content": "帮我看看ID为1的用户的角色是什么，顺便算一下目前8核处理器，12个线程活跃情况下的系统负载。回答时态度友好一点。"}]

            print("🗣️ 正在请求 OpenAI 思考...")

            # 5. 第一次调用大模型，触发 Function Calling
            response = await llm_client.chat.completions.create(
                model="gpt-4o",
                messages=messages,
                tools=openai_tools
            )

            message = response.choices[0].message
            messages.append(message) # 将助手的回复加入历史

            # 6. 处理模型返回的工具调用请求
            if message.tool_calls:
                for tool_call in message.tool_calls:
                    tool_name = tool_call.function.name
                    tool_args = json.loads(tool_call.function.arguments)

                    print(f"🤖 LLM 决定调用工具: {tool_name}({tool_args})")

                    # 7. 使用 MCP Client 执行真实逻辑
                    mcp_result = await session.call_tool(tool_name, tool_args)
                    result_text = mcp_result.content[0].text
                    print(f"✅ 工具执行结果: {result_text}")

                    # 8. 将结果封装成 tool message 喂回给大模型
                    messages.append({
                        "role": "tool",
                        "tool_call_id": tool_call.id,
                        "name": tool_name,
                        "content": result_text
                    })

                # 9. 第二次调用大模型，生成最终回复
                print("🗣️ 正在请求 OpenAI 总结...")
                final_response = await llm_client.chat.completions.create(
                    model="gpt-4o",
                    messages=messages
                )
                print(f"\n🎉 最终回答:\n{final_response.choices[0].message.content}")

if __name__ == "__main__":
    asyncio.run(llm_mcp_agent())
```

#### 业界更通用的是哪种？
毫无疑问，方式二（结合大模型 Function Calling）是目前的绝对主流和核心应用场景。

__为什么？__

1. MCP 的诞生使命：MCP 协议被设计出来的初衷，就是为了解决“如何让大模型标准化地访问外部世界”这个问题。如果没有大模型的大脑，MCP 充其量只是一个比较麻烦的微服务通信协议（远不如 gRPC 或 HTTP RESTful 好用）。

2. 解耦的极致：通过方式二，后端开发工程师只需要专心写 Python/Java 服务端逻辑（通过 FastMCP 等暴露出去），前端/Agent 工程师只需要专心调 Prompt 和编排工作流。

- MCP Server 成了四肢。

- OpenAI/Claude 成了大脑。

- MCP Client 成了连接两者的神经中枢。

__业界的实际工程形态__

在实际的生产环境中，你甚至不需要像我上面演示的那样手写 for tool in mcp_tools_response.tools: 去做格式转换。
